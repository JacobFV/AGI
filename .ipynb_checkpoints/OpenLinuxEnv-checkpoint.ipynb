# OpenLinuxEnv

A simple reward-free python interface to VNC environments for agents to interact in. Your local desktop, a remote server, or a container service can host docker containers with VNC servers. 

## Getting Started
The starter instructions use the `dorowu/ubuntu-desktop-lxde-vnc` docker image because of [this answer](https://github.com/fcwu/docker-ubuntu-vnc-desktop) and [that answer](https://stackoverflow.com/questions/40658095/how-to-open-ubuntu-gui-inside-a-docker-image), but any container or VM with VNC support should do the trick. Try spinning up an A2 Google [compute engine instance](https://medium.com/@piyushraj246800/your-guide-to-vnc-server-on-gcp-ed8255c81b43) or use multiple nvidia-docker instances of [ikeyasu/reinforcement-learning](https://github.com/ikeyasu/docker-reinforcement-learning) to train deep learning with deep learning! (Slower but cheaper option: train the agent to explore [colab.research.google.com]()) 

0. **Make sure you have a VNC server running on your target system.** 
    - *If you want to run a VNC server locally*:
        1. [Install docker](https://docs.docker.com/get-docker/)
        2. Run this terminal command `docker run --name ubvnc -p 6080:80 -p 5900:5900 -v /dev/shm:/dev/shm dorowu/ubuntu-desktop-lxde-vnc`
            - When you're ready to stop the container, enter `Ctrl-C` in the terminal. Then you should be able to start and stop the container in the future by `docker start ubvnc` and `docker stop ubvnc`.
        3. You can make sure any VNC server is working by using a standard VNC client like tigervnc, novnc, realvnc, etc.
    - *If the target is not your local machine*, ensure firewall permissions between the `OpenLinuxEnv` instance and VNC instance are configured to not block the IP and port addresses used to communicate. Obviously, use good security practices. 
1. Install pyVNC. If you're using a package manager like conda, make sure pyVNC is installed in the same environment where OpenVNCEnv will be installed. Enter the following terminal commands:
    ```bash
    git clone https://github.com/cair/pyVNC
    cd pyVNC
    pip install .
    ```
2. **Install OpenVNCEnv**: `pip install openvncenv`
3. **Test OpenVNCEnv**.
    1. Launch a VNC server (see step 0)
    2. Run the test script: `python -m openvncenv.test --ip IP --port PORT`
    3. If you recieve the message `OpenVNCEnv sucessfully commicated with the target`, you should be good to go! Otherwise, see [TROUBLESHOOTING](link to troubleshooting section in reading or on github)

````python
import pyVNC
from pyVNC.Client import Client

import numpy as np
import matplotlib.pyplot as plt
````

````python
vnc = Client(host='127.0.0.1',
             password=None,
             port=5900,
             depth=32,
             fast=False,
             shared=True)
vnc.start()
s = vnc.screen # Get a array representation of the screen shape: (?, ?, 3)
plt.imshow(np.array(s.get_array()))
````
